2. **Functions**-organizing your code better
Function: A block of code designed to perform a specific task without duplication of code.
   

```python
#Example
# pay = no.of hours * pay per hour
# 12% tax deductions/ 12 % tax bracket
def get_pay(num_hours):
    pay_preTax=num_hours*15
    pay_afterTax=pay_preTax*(1-0.12)
    return pay_afterTax
pay_full=get_pay(40)
print(pay_full)

```

```python
#scope : var inside a body has its scope of access or usage only inside that same body
print(pay_afterTax)
```

```python
#multiple arguments
def get_pay_with_more_inputs(num_hours, hourly_wage, tax_bracket):
    pay_preTax=num_hours*hourly_wage
    pay_afterTax=pay_preTax*(1-tax_bracket)
    return pay_afterTax
higher_pay_afterTax=get_pay_with_more_inputs(56,40,.32)
print(higher_pay_afterTax)    
```

```python
#no arguments
def print_hello():
    print("Kaggle")
    print("Python")
    print("Jupyter")
print_hello()
```

```python
def get_expected_cost(beds, baths):
    fixed_cost_house=80000
    total_bed_cost=beds*30000
    total_bath_cost=baths*10000
    total_exp_cost=fixed_cost_house+total_bed_cost+total_bath_cost
    return total_exp_cost #used var for better understanding 
    #80000+(beds*30000)+(baths*10000) also works
get_expected_cost(2,1)

    
```

```python
op1=get_expected_cost(2,3)
op2=get_expected_cost(3,2)
op3=get_expected_cost(3,3)
op4=get_expected_cost(3,4)
print(op1)
print(op2)
print(op3)
print(op4)
```

```python
def get_cost(sqft_walls, sqft_ceiling, sqft_per_gallon, cost_per_gallon):
    #total area/area covered by one gallon of paint = no.of gallons req
    cost=cost_per_gallon*((sqft_walls+sqft_per_gallon)/sqft_per_gallon)#doesn't round off, '/' division for floating point 
    #'//' division for int values only
    return cost
get_cost(40,50,3,20)
    
```

```python
project_cost=get_cost(432,144,400,15)
print(project_cost)
```

```python
#math.ceil() #valid to use function inside a function
import math
def get_actual_cost(walls, ceiling, sqft_per_gallon, cost_per_gallon):
    cost=cost_per_gallon*(math.ceil((walls+ceiling)/sqft_per_gallon))
    return cost
get_actual_cost(456, 78,23,16)
```

```python
def cost_of_project(engraving, solid_gold):
    if solid_gold==True:
        base=100
        eng=10
    else:
        base=50
        eng=7
    cost=base+(eng*len(engraving))
    return cost
print(cost_of_project("python-learn ing",True))
print(cost_of_project("python ", False))
```

The Idea is to USE " not "

```python
def cost_project(engraving, solid_gold):
    cost= solid_gold * (100+10*len(engraving)) + (not solid_gold) * (50+7*len(engraving))
    return cost
print(cost_project("06/03/2026", False))
```
